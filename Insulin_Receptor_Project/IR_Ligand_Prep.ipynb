```python
# Step 1: Install RDKit (if not already installed)
!pip install rdkit

# Step 2: Import necessary libraries
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.MolStandardize import rdMolStandardize

# --- THE FIX IS HERE ---
# We replaced the long wrong code with the correct Indole-Butyl-Amine SMILES
correct_smiles = "C1=CC=C2C(=C1)C(=CN2)CCCCN"
print(f"Processing Ligand: {correct_smiles}")

# Step 3: Create the Molecule object
mol = Chem.MolFromSmiles(correct_smiles)

# Step 4: Standardization (Neutralize charges and clean)
# This removes salts and fixes weird chemistry errors
clean_mol = rdMolStandardize.Cleanup(mol)
uncharger = rdMolStandardize.Uncharger()
mol_neutral = uncharger.uncharge(clean_mol)
mol_neutral = Chem.AddHs(mol_neutral) # Add Hydrogens (Crucial for 3D)

# Step 5: Generate 3D Conformer using ETKDG
# This turns the flat text into a 3D shape with depth
params = AllChem.ETKDG()
AllChem.EmbedMolecule(mol_neutral, params)

# Step 6: Save the 3D model as a PDB file
output_filename = "Indole_Correct_Ligand.pdb"
Chem.MolToPDBFile(mol_neutral, output_filename)

print(f"SUCCESS: 'Step 7' is Complete.")
print(f"Your 3D file is saved as: {output_filename}")
```
